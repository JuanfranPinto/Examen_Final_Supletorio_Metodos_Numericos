```python
import numpy as np
import matplotlib.pyplot as plt
from src import gauss_jacobi, gauss_seidel
```

```python
def trayectoria_gs(A, b, x0, tol=1e-6, it=10):
    # Inicializamos la trayectoria con el punto inicial
    tray = [x0.copy().reshape(-1, 1)]
    x_actual = x0.copy()
    
    for k in range(it):
        # NOTA: Cambiamos 'x0' por 'x' para que coincida con tu librería
        x_actual = gauss_seidel(A=A, b=b, x=x_actual, tol=tol, max_iter=1)
        tray.append(x_actual)
        
    return np.hstack(tray).T
```

```python
def dibujar(tray, titulo):
    x = tray[:,0]
    y = tray[:,1]

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, "o-", color="blue", alpha=0.6) # Línea con puntos
    plt.plot(x[0], y[0], "gd", markersize=10, label="X0 (Inicio)")
    plt.plot(x[-1], y[-1], "ro", markersize=10, label="X* (Final)")
    plt.xlabel("Eje X")
    plt.ylabel("Eje Y")
    plt.title(titulo)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```python
A1 = np.array([[2, 3], [1, 2]], float)
b1 = np.array([5, 8], float)
x0_1 = np.array([0, 0], float)

tray1 = trayectoria_gs(A1, b1, x0_1)
dibujar(tray1, "Primer Ejercicio: Trayectoria Gauss-Seidel")

sol1 = tray1[-1]
# Pendiente m = (y2-y1)/(x2-x1)
m1 = (tray1[-1, 1] - tray1[0, 1]) / (tray1[-1, 0] - tray1[0, 0])

print(f"--- Ejercicio 1 ---")
print("Respuesta (Solución):", sol1)
print("Pendiente m =", m1)
```
